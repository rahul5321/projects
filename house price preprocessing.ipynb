```python
def col_dict(data):
    cols={'drop_cols':[],
      'null_cols':[],
     'non_null_cols':[],'num_cols':[],'cat_cols':[],'null_num_cols':[],'null_cat_cols':[]}
    size=data.shape[0]
    for col in data.columns:
        null=data[col].isnull().sum()
        dtype=(data[col].dtypes=='object')
        if null>0.3*size:
            cols['drop_cols'].append(col)
        elif null<0.3*size and null!=0:
            cols['null_cols'].append(col)
            if dtype:
                cols['null_cat_cols'].append(col)
            else:
                cols['null_num_cols'].append(col)
        else:
            cols['non_null_cols'].append(col)
    
    for col in data.columns:
        null=data[col].isnull().sum()
        dtype=(data[col].dtypes=='object')
        if dtype and null<0.3*size:
            cols['cat_cols'].append(col)
        elif not dtype and null<0.3*size:
            cols['num_cols'].append(col)
    return cols
```

```python
#funcn for droppping column input-dataframe and dict
def dropping(df,column):
    df=df.drop(column['drop_cols'],axis=1)
    return df
data=dropping(data,cols)
data.shape
```

```python
#fiiling missing values, input- dataframe and dict
def impute(df,column):
    for col in column['num_cols']:
        df[col]=df[col].fillna(df[col].median())
    for col in column['cat_cols']:
        df[col]=df[col].fillna(df[col].mode()[0])
    return df
    #df[column['num_cols']]=df[column['num_cols']].fillna(df[column['num_cols']].median())
    #df[column['cat_cols']]=df[column['cat_cols']].fillna(df[column['cat_cols']].mode())
data=impute(data,cols)
print(sum(data.isnull().sum()==0))
    
    
```
